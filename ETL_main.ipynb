```python
# Import Dependencies
import pandas as pd
import requests
from pprint import pprint
from config import keys
```

```python
# Define URL and API key
url = 'https://api.sportsdata.io/v3/nba/stats/json/PlayerSeasonStats/2022?key={}'.format(keys)
```

```python
print(url)
```

```python
# Send GET request to nba API
response = requests.get(url)

# Parse response as JSON
data = response.json()
# data = data[0]

# grep keys from data
data_columns = [keys for keys in data[0]]
```

```python
df = pd.read_json(url)
df
```

```python
df.columns
```

```python
# check for missing data
df.isna().sum()
```

```python
# create a copy pf dataframe
df_copy = df.copy()
```

```python
# Set playerID as Index
df_copy = df_copy.set_index("Name")
```

```python
# dropped columns not needed
df_copy = df_copy.drop(columns = ['LineupConfirmed', 'FantasyPoints',
                                  'StatID','Updated',
                                  'LineupStatus',
                                  'Season','GlobalTeamID',                                  
                                  'IsClosed',
                                  'FantasyPointsFantasyDraft','FantasyPointsFanDuel',
                                  'FantasyPointsDraftKings', 'FantasyPointsYahoo','SeasonType','TeamID', 'PlayerID' ], axis =1)
```

```python
df_copy.columns
```

```python
df_dummies = pd.get_dummies(df_copy)
```

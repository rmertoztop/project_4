```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from pyspark.sql.functions import col
import numpy as np
import seaborn as sns
```

```python
# Select the features (X) and target variables (y) for points (PTS) and assists (AST)
features = joined_data[['Age_x', 'G_x', 'GS_x', 'MP_x', 'FG_x', 'FGA_x',
                       'Age_y', 'G_y', 'GS_y', 'MP_y', 'FG_y', 'FGA_y']]

target_pts = joined_data[['PTS_x']]

target_ast = joined_data[['AST_x']]

```

```python
# Split the data into training and test sets
X_train, X_test, y_train_pts, y_test_pts, y_train_ast, y_test_ast = train_test_split(features, target_pts, target_ast, test_size=0.4, random_state=42)
```

# Data Transformations

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Make a copy of Dataset
joined_data_copy = joined_data.copy()
```

```python
# Drop player column
joined_data_copy = joined_data_copy.drop('Player', axis =1)
joined_data_copy
```

```python
#CHeck types of Data
joined_data_copy.dtypes
```

```python
# Scale only numeric dataset
numerical = joined_data_copy.select_dtypes(include='float64').columns
joined_data_copy.loc[:,numerical]= StandardScaler().fit_transform(joined_data_copy.loc[:,numerical])
```

```python
# Create a copy of transformed data
joined_data_copy2 = joined_data_copy.copy()
```

```python
# Drop Position and team columns
joined_data_copy = joined_data_copy.drop(["Pos_x","Tm_x","Pos_y","Tm_y"], axis=1)
joined_data_copy.shape
```

```python
# Transform the Position and Team column using get_dummies()
dummies = pd.get_dummies(joined_data_copy2[["Pos_x","Tm_x","Pos_y","Tm_y"]])
# Display sample data
dummies.head()
```

```python
# Concatenate the df_shopping_transformed and the card_dummies DataFrames
df_transformed = pd.concat([joined_data_copy, dummies], axis=1)

# Display sample data
df_transformed.head()
```

```python
# define the features (X)
features = df_transformed.drop(['PTS_x', 'AST_x'], axis =1)

```

```python
# Define target vector
target_pts = df_transformed['PTS_x'].values.reshape(-1, 1)

# target_ast = df_transformed['AST_x'].values.reshape(-1, 1)
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target_pts, random_state=42)
```

```python
# Create the regression instance
model_pts = LinearRegression()

model_ast = LinearRegression()

# Fitting the model
model_pts.fit(X_train, y_train)

model_ast.fit(X_train, y_train)
```

```python
# Make predictions for points
pts_predictions = model_pts.predict(X_test)

# ast_predictions = model_ast.predict(X_test)
```

# Model Accuracy

```python
# Model Accuracy for Points
print('Test Acc: %.3f' % model_pts.score(X_test, y_test))
# MSE
print('MSE: %.3f' %np.sqrt(metrics.mean_squared_error(y_test, pts_predictions)))
```
